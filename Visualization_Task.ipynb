```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
# histogram
sns.histplot(data=penguins,x='flipper_length_mm',kde=True)
plt.title("Flipper Length")
plt.show()
```

Here in this graph most penguins have flipper length between 190 - 195mm, a very few have flipper length around 170 and 230mm.The flipper length are not evenly distributed.

```python
# scatterplot
sns.scatterplot(data=penguins,x='bill_length_mm',y='bill_depth_mm',hue='species')
```

Here the species Adelie have large bill depth and short bill length,
species Chinstrap have large bill legth and large bill depth,
species Gentoo have short bill depth and largest bill length.

```python
# countplot
sns.countplot(data=penguins,x='island')
```

Here most penguins are from Biscoe island, second highest number of penguins from Dream and least from Torgersen.

```python
# correlation
corr = penguins.corr(numeric_only=True)
corr
```

```python
# heatmap
sns.heatmap(corr,annot=True,cmap='Blues')
```

Here strongest positive correlation is between flipper length and body mass which indicates penguins with longer flipper lengths will be heavier.Negative correlation is between bill depth and flipper length.

```python
# violin plot
sns.violinplot(data=penguins,x='species',y='body_mass_g')
```

Here Gentoo penguins have the highest body mass, Adelie and Chinstarp have lower and similar body masses.

```python
# strip plot
sns.stripplot(data=penguins,x='species',y='bill_length_mm')
plt.title("Strip plot")
```

Here Adelie penguins have the shortest bills,Chinstrap and Gentoo have the long bills.

```python
# pie chart
species_count = penguins['species'].value_counts()
plt.pie(species_count,labels=species_count.index,shadow=True,autopct='%1.1f%%')
plt.show()
```

Here large number of penguins are Adelie species about 44.2%,Gentoo is the second largest and least are Chinstrap.

```python
# bar chart
sns.barplot(data=penguins,x='sex',y='body_mass_g',color='yellow')
```

Here male penguins have high body mass than female penguins around 4500g.



```python
# box plot
sns.boxplot(data=penguins,x='species',y='body_mass_g',hue='sex')
```

Here male penguins have a higher body mass than female penguins in all three species.Gentoo penguins have the highest body mass among all species.

```python
# stem plot
plt.stem(penguins['bill_length_mm'],penguins['bill_depth_mm'])
plt.title("Stem plot")
plt.xlabel("Bill Length")
plt.ylabel("Bill Depth")
```

Here the bill length and bill depth varies.
